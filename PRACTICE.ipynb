```python
def add(numbers):
    return sum(numbers)

def sub(numbers):
    result = numbers[0]
    for i in range(1, len(numbers)):
        result -= numbers[i]
    return result

def mul(numbers):
    result = 1
    for num in numbers:
        result *= num
    return result

def div(numbers):
    result = numbers[0]
    for i in range(1, len(numbers)):
        result /= numbers[i]
    return result

while True:
    print('1. Addition')
    print('2. Subtraction')
    print('3. Multiplication')
    print('4. Division')
    print('5. Exit')

    num_count = int(input('Enter the number of values: '))
    numbers = []
    for i in range(num_count):
        number = float(input(f'Enter number {i+1}: '))
        numbers.append(number)

    choice = int(input('Enter your choice: '))

    if choice == 5:
        print('Exiting the program...')
        break

    elif choice == 1:
        result = add(numbers)
        for num in numbers:
            print("The numbers are:", num)
        print("\nThe sum of the numbers is:", result)
    elif choice == 2:
        result = sub(numbers)
        for num in numbers:
            print("The numbers are:", num)
        print("\nThe result of subtraction is:", result)
    elif choice == 3:
        result = mul(numbers)
        for num in numbers:
            print("The numbers are:", num)
        print("\nThe result of multiplication is:", result)
    elif choice == 4:
        result = div(numbers)
        for num in numbers:
            print("The numbers are:", num)
        print("\nThe result of division is:", result)
    else:
        for num in numbers:
            print("The numbers are:", num)
        print("\nInvalid choice! Please enter a valid option.")
```

```python
import string

class URLShortener:
    BASE62 = string.digits + string.ascii_letters
    BASE = len(BASE62)
    urls = {}
    counter = 1

    @classmethod
    def encode(cls, num):
        """Encode a given number into base62"""
        result = []
        while num > 0:
            num, rem = divmod(num, cls.BASE)
            result.append(cls.BASE62[rem])
        return ''.join(reversed(result))

    @classmethod
    def shorten_url(cls, url):
        if url in cls.urls:
            short_url = cls.urls[url]
        else:
            short_url = cls.encode(cls.counter)
            cls.urls[url] = short_url
            cls.counter += 1
        return short_url

# Create an instance of the URLShortener class
shortener = URLShortener()

# Ask the user for a URL
url = input("Enter the URL to shorten: ")

# Shorten the URL
short_url = shortener.shorten_url(url)

# Display the shortened URL
print("Shortened URL:", short_url)

# Increment the counter for the next run
shortener.counter += 1
```

```python
import random

def encode_message(message):
    words = message.split()
    encoded_words = []

    for word in words:
        if len(word) >= 3:
            word = word[1:] + word[0]  # Remove first letter and append it at the end
            word = random_chars(3) + word + random_chars(3)  # Add three random characters at the beginning and end
        else:
            word = word[::-1]  # Reverse the string
        encoded_words.append(word)

    encoded_message = ' '.join(encoded_words)
    return encoded_message

def decode_message(message):
    words = message.split()
    decoded_words = []

    for word in words:
        if len(word) < 3:
            word = word[::-1]  # Reverse the string
        else:
            word = word[3:-3] + word[-4]  # Remove three random characters from start and end, then move the last letter to the beginning
        decoded_words.append(word)

    decoded_message = ' '.join(decoded_words)
    return decoded_message

def random_chars(length):
    characters = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    random_chars = ''.join(random.choice(characters) for _ in range(length))
    return random_chars

choice = input("Do you want to code (C) or decode (D) a message? ").upper()

if choice == 'C':
    message = input("Enter the message you want to encode: ")
    encoded_message = encode_message(message)
    print("Encoded message:", encoded_message)
elif choice == 'D':
    message = input("Enter the message you want to decode: ")
    decoded_message = decode_message(message)
    print("Decoded message:", decoded_message)
else:
    print("Invalid choice. Please enter 'C' to code or 'D' to decode a message.")
```

```python
import random

def encode_word(word):
    if len(word) >= 3:
        encoded_word = word[1:] + word[0] + ''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=3))
        encoded_word += ''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=3))
    else:
        encoded_word = word[::-1]
    return encoded_word

def decode_word(word):
    if len(word) < 3:
        decoded_word = word[::-1]
    else:
        decoded_word = word[3:-3]
        decoded_word = decoded_word[-1] + decoded_word[:-1]
    return decoded_word

def encode_message(message):
    words = message.split()
    encoded_message = ' '.join(encode_word(word) for word in words)
    return encoded_message

def decode_message(message):
    words = message.split()
    decoded_message = ' '.join(decode_word(word) for word in words)
    return decoded_message

def main():
    choice = input("Enter '1' to encode or '2' to decode: ")
    if choice == '1':
        message = input("Enter the message to encode: ")
        encoded_message = encode_message(message)
        print("Encoded message:", encoded_message)
    elif choice == '2':
        message = input("Enter the message to decode: ")
        decoded_message = decode_message(message)
        print("Decoded message:", decoded_message)
    else:
        print("Invalid choice!")

if __name__ == '__main__':
    main()
```
