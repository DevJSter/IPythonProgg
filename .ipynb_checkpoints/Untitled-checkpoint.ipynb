```python
def main():
    k0 = [0] * 10
    k1 = []
    k2 = []

    nums = int(input("How Many Numbers you want to enter: "))
    for i in range(nums):
        k0[i] = int(input(f"Enter number {i + 1}:"))

    num_clusters = int(input("\nEnter the number of clusters to be created:"))

    print("\nEnter initial mean 1:")
    m1 = int(input())
    print("\nEnter initial mean 2:")
    m2 = int(input())

    om1, om2 = 0, 0  # Old means

    while True:
        om1 = m1
        om2 = m2

        k1.clear()
        k2.clear()

        for i1 in range(10):
            t1 = abs(k0[i1] - m1)
            t2 = abs(k0[i1] - m2)

            if t1 < t2:
                k1.append(k0[i1])
            else:
                k2.append(k0[i1])

        if len(k1) > 0:
            m1 = sum(k1) / len(k1)
        if len(k2) > 0:
            m2 = sum(k2) / len(k2)

        print("\nCluster 1:", k1)
        print("Mean 1:", m1)
        print("\nCluster 2:", k2)
        print("Mean 2:", m2)
        print("\n----")

        if abs(m1 - om1) < 0.0001 and abs(m2 - om2) < 0.0001:
            break

    print("\nClusters created")

if __name__ == "__main__":
    main()
```
