```python
from pytube import YouTube
from tqdm import tqdm
import os
from moviepy.editor import VideoFileClip, AudioFileClip

input("THIS IS FOR EDUCATIONAL PURPOSES ONLY. I AM NOT RESPONSIBLE FOR WHAT YOU DO WITH THIS PROGRAM (press enter to continue)")

place = "Where do you want to save the file: "
choice = input("Do you want to download (1. A single video or (2. Multiple videos: ")

if choice == '1':
    save = input(place)
    while True:
        URL = input("What is the link: ")
        try:
            yt = YouTube(URL)
        except Exception as e:
            print("Error: ", str(e))
            continue

        print("Author: ", yt.author)
        print("Published: ", yt.publish_date)
        print("Title: ", yt.title)
        print("Views: ", yt.views)
        print("Description: ", yt.description)
        print("Length (seconds): ", yt.length)
        print("Rating: ", yt.rating)

        available_streams = yt.streams.filter(progressive=True, file_extension="mp4")
        if not available_streams:
            print("No valid video streams found for this video.")
        else:
            print("Available Video Formats:")
            for idx, stream in enumerate(available_streams):
                print(f"{idx + 1}. {stream.resolution} - {stream.filesize / (1024 * 1024):.2f} MB")

            selected_stream_index = int(input("Enter the number corresponding to the desired video quality: ")) - 1

            if 0 <= selected_stream_index < len(available_streams):
                yd = available_streams[selected_stream_index]
                print("DOWNLOADING...")
                yd.download(save)
                print("DOWNLOAD COMPLETE")
            else:
                print("Invalid selection. Please choose a valid number.")

elif choice == '2':
    save2 = input(place)
    file = input("What is the file containing the links: ")
    while True:
        with open(file, 'r') as link:
            for i in tqdm(link, desc="Downloading"):
                try:
                    yt = YouTube(i)
                except Exception as e:
                    print("Error: ", str(e))
                    continue

                video_stream = yt.streams.filter(file_extension='mp4', res='720p', progressive=True).first()
                audio_stream = yt.streams.filter(only_audio=True).first()

                video_file = save2 + "/video.mp4"
                audio_file = save2 + "/audio.mp3"

                print("Downloading video...")
                video_stream.download(output_path=save2, filename='video')
                print("Downloading audio...")
                audio_stream.download(output_path=save2, filename='audio')

                print("Merging video and audio...")
                video = VideoFileClip(video_file)
                audio = AudioFileClip(audio_file)

                final_output = save2 + "/output.mp4"
                final_video = video.set_audio(audio)
                final_video.write_videofile(final_output)

                # Delete temporary video and audio files
                os.remove(video_file)
                os.remove(audio_file)

    print("FINISHED")
```

```python
from pytube import YouTube
from tqdm import tqdm
import os

input("THIS IS FOR EDUCATIONAL PURPOSES ONLY. I AM NOT RESPONSIBLE FOR WHAT YOU DO WITH THIS PROGRAM (press enter to continue)")

place = "Where do you want to save the file: "
choice = input("Do you want to download (1. Video, 2. Audio, 3. Both): ")

if choice == '1':
    save = input(place)
    while True:
        URL = input("What is the link: ")
        try:
            yt = YouTube(URL)
        except Exception as e:
            print("Error: ", str(e))
            continue

        print("Author: ", yt.author)
        print("Published: ", yt.publish_date)
        print("Title: ", yt.title)
        print("Views: ", yt.views)
        print("Description: ", yt.description)
        print("Length (seconds): ", yt.length)
        print("Rating: ", yt.rating)

        available_streams = yt.streams.filter(progressive=True, file_extension="mp4")
        if not available_streams:
            print("No valid video streams found for this video.")
        else:
            print("Available Video Formats:")
            for idx, stream in enumerate(available_streams):
                print(f"{idx + 1}. {stream.resolution} - {stream.filesize / (1024 * 1024):.2f} MB")

            selected_stream_index = int(input("Enter the number corresponding to the desired video quality: ")) - 1

            if 0 <= selected_stream_index < len(available_streams):
                yd = available_streams[selected_stream_index]
                print("DOWNLOADING...")
                yd.download(save)
                print("DOWNLOAD COMPLETE")
            else:
                print("Invalid selection. Please choose a valid number.")

elif choice == '2':
    save = input(place)
    while True:
        URL = input("What is the link: ")
        try:
            yt = YouTube(URL)
        except Exception as e:
            print("Error: ", str(e))
            continue

        audio_streams = yt.streams.filter(only_audio=True)
        if not audio_streams:
            print("No valid audio streams found for this video.")
        else:
            print("Available Audio Formats:")
            for idx, stream in enumerate(audio_streams):
                print(f"{idx + 1}. {stream.abr}kbps")
            
            selected_stream_index = int(input("Enter the number corresponding to the desired audio quality: ")) - 1

            if 0 <= selected_stream_index < len(audio_streams):
                audio_stream = audio_streams[selected_stream_index]
                print("DOWNLOADING...")
                
                # Save audio with video title as filename
                video_title = yt.title
                audio_filename = f"{video_title}.mp3"
                audio_stream.download(save, filename=audio_filename)
                
                print("DOWNLOAD COMPLETE")
            else:
                print("Invalid selection. Please choose a valid number.")

elif choice == '3':
    save = input(place)
    while True:
        URL = input("What is the link: ")
        try:
            yt = YouTube(URL)
        except Exception as e:
            print("Error: ", str(e))
            continue

        print("Downloading Video...")
        video_stream = yt.streams.filter(progressive=True, file_extension="mp4").first()
        video_filename = f"{yt.title}.mp4"
        video_stream.download(save, filename=video_filename)
        print("Video Download Complete")

        print("Downloading Audio...")
        audio_stream = yt.streams.filter(only_audio=True).first()
        audio_filename = f"{yt.title}.mp3"
        audio_stream.download(save, filename=audio_filename)
        print("Audio Download Complete")
        
        

    print("Download Complete")
```
