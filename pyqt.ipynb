```python
import random
import PySimpleGUI as sg

def mad_libs():
    # Define a list of stories to choose from
        stories = [
        {
            'title': 'The Haunted House',
            'template': 'One dark and {adjective} night, a group of {noun} decided to explore the abandoned {place}. \n As they entered the {place}, they heard a {sound} coming from upstairs.\n They {verb} up the stairs, only to find a {adjective} ghost waiting for them. \nThey {verb} in terror and never returned to the {place} again.'
        },
        {
            'title': 'The Camping Trip',
            'template': 'Last summer, my family and I went on a camping trip to {place}. \nWe spent our days hiking in the {adjective} mountains and swimming in the {noun}. \nAt night, we would sit around the campfire and {verb} stories.\n One night, we heard a {sound} coming from the woods.\n We {verb} to investigate, but it turned out to be just a {adjective} {noun}.\n We all {verb} in relief and went back to our campsite.'
        },
        {
            'title': 'The Secret Garden',
            'template': 'One bright and beautiful morning, a group of curious children decided to explore the hidden garden behind the old {place}. As they entered the garden, they heard a {sound} coming from the overgrown bushes. They {verb} closer, only to find a {adjective} fairy waiting for them. They {verb} in amazement as the fairy granted them each a wish before disappearing. The children never revealed the location of the secret garden and continued to visit it in secret.'
        },
        {
            'title': 'The Ski Trip',
            'template': 'Last winter, my friends and I went on a ski trip to {place}. We spent our days skiing down the {adjective} slopes and relaxing in the {noun}. At night, we would {verb} by the fireplace and share stories. One night, we heard a {sound} coming from outside the cabin. We {verb} to investigate, but it turned out to be just a {adjective} snowman. We all {verb} in laughter and went back inside to enjoy the rest of our trip.'
        },
            {
            'title': 'The Abandoned Warehouse',
            'template': 'One stormy and {adjective} night, a group of teenagers decided to explore the abandoned warehouse on the outskirts of town. As they entered the warehouse, they heard a {sound} coming from deep within. They {verb} cautiously, only to find a {adjective} robot waiting for them. The robot had been left behind by its creators and was in need of repair. The teenagers {verb} the robot and brought it back to life, becoming unlikely friends with the machine.'
        },
        {
            'title': 'The Beach Vacation',
            'template': 'Last summer, my family and I went on a beach vacation to {place}. We spent our days playing in the {noun} and lounging under the {adjective} sun. At night, we would {verb} by the bonfire and share stories. One night, we heard a {sound} coming from the water. We {verb} to investigate, but it turned out to be just a {adjective} dolphin swimming nearby. We all {verb} in awe and watched the dolphin swim away into the night.'
        {
            'title': 'The Mysterious Mansion',
            'template': 'One foggy and {adjective} evening, a group of detectives decided to investigate the mysterious mansion on the hill. As they entered the mansion, they heard a {sound} coming from the basement. They {verb} down the stairs, only to find a {adjective} treasure waiting for them. The detectives {verb} in excitement as they uncovered the long-lost treasure and solved the mystery of the mansion.'
        },
    
        {
            'title': 'The Beach Vacation',
            'template': 'Last summer, my family and I went on a beach vacation to Hawaii. We spent our days lounging on the sandy beaches and snorkeling in the crystal clear waters. At night, we would sit around the campfire and share stories. One night, we heard a strange noise coming from the nearby jungle. We ventured into the jungle to investigate, but it turned out to be just a playful monkey swinging from the trees. We all laughed in relief and went back to our beach house.'
        }
  ]
            
    

    # Define the layout of the input fields and generate button
    layout = [
        [sg.Text('Adjective:'), sg.InputText(key='adjective')],
        [sg.Text('Noun:'), sg.InputText(key='noun')],
        [sg.Text('Verb:'), sg.InputText(key='verb')],
        [sg.Text('Place:'), sg.InputText(key='place')],
        [sg.Text('Sound:'), sg.InputText(key='sound')],
        [sg.Button('Generate Mad Lib', key='generate_button')],
        [sg.Text('', key='result_label')]
    ]

    # Create the window
    window = sg.Window('Mad Libs', layout)

    # Start the event loop
    while True:
        event, values = window.read()
        if event == sg.WINDOW_CLOSED:
            break
        elif event == 'generate_button':
            # Choose a random story from the list
            story = random.choice(stories)

            # Get the user's inputs
            adjective = values['adjective']
            noun = values['noun']
            verb = values['verb']
            place = values['place']
            sound = values['sound']

            # Generate the story using the user's inputs
            mad_lib = story['template'].format(adjective=adjective, noun=noun, verb=verb, place=place, sound=sound)

            # Update the label to display the completed Mad Lib
            window['result_label'].update(f"{story['title']}\n\n{mad_lib}")

    # Close the window
    window.close()

# Call the mad_libs function to run the program
mad_libs()
```
