```python
# Required libraries
import pyttsx3
import speech_recognition as sr
import datetime
import wikipedia
import webbrowser
import os
import smtplib
import pywhatkit
import random
import requests
import json
from pprint import pprint

# Initialize text-to-speech engine
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)


def speak(audio):
    """Speak the provided audio string using the text-to-speech engine"""
    engine.say(audio)
    engine.runAndWait()


def wish_me():
    """Greet the user based on the current time"""
    hour = datetime.datetime.now().hour
    if 0 <= hour < 12:
        speak("Good morning!")
    elif 12 <= hour < 18:
        speak("Good afternoon!")
    else:
        speak("Good evening!")
    speak("I am your personal assistant. How can I assist you?")


def take_command():
    """Use speech recognition to get the user's command"""
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source)
    try:
        print("Recognizing...")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n")
    except Exception as e:
        print(e)
        print("Sorry, I did not understand. Could you please repeat?")
        return None
    return query.lower()


def open_website(url):
    """Open the provided website in the default web browser"""
    webbrowser.open(url)


def play_song(song_name):
    """Play the provided song on YouTube"""
    pywhatkit.playonyt(song_name)


def get_weather(city_name):
    """Get the current weather of the provided city"""
    weather_api_key = "YOUR_API_KEY"
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    complete_url = f"{base_url}appid={weather_api_key}&q={city_name}"
    response = requests.get(complete_url)
    weather_data = response.json()
    if weather_data["cod"] != "404":
        weather_description = weather_data["weather"][0]["description"]
        temperature = weather_data["main"]["temp"]
        feels_like = weather_data["main"]["feels_like"]
        speak(f"The weather in {city_name} is currently {weather_description}. "
              f"The temperature is {temperature} Kelvin, but it feels like {feels_like} Kelvin.")
    else:
        speak("Sorry, I could not retrieve the weather information for that city.")


def send_email(to_address, message):
    """Send an email to the provided email address"""
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login('YOUR_EMAIL_ADDRESS', 'YOUR_EMAIL_PASSWORD')
    server.sendmail('YOUR_EMAIL_ADDRESS', to_address, message)
    server.quit()


def get_joke():
    """Get a random joke"""
    joke_api_url = "https://official-joke-api.appspot.com/random_joke"
    response = requests.get(joke_api_url)
    joke_data = json.loads(response.text)
    setup = joke_data["setup"]
    punchline = joke_data["punchline"]
    speak(setup)
    speak(punchline)


def main():
    """Main program loop"""
    wish_me()
    while True:
        command = take_command()
        if not command:
            continue
        elif "open google" in command:
            open_website("https://www.google.com/")
        elif "play" in command:
            song_name = command.replace("play", "")
            speak(f"Playing {song_name} on YouTube")
            play_song(song_name)
        elif "weather" in command:
            city_name = command.replace("weather in", "")
            get_weather(city_name)
        elif "send email" in command:
            try:
                speak("What should I say?")
                message = take_command()
                to_address = "example@example.com"  # Replace with recipient's email address
                send_email(to_address, message)
                speak("Email has been sent!")
            except Exception as e:
                print(e)
                speak("Sorry, I was not able to send the email.")
        elif "joke" in command:
            get_joke()
        elif "time" in command:
            time_str = datetime.datetime.now().strftime("%H:%M:%S")
            speak(f"The current time is {time_str}")
        elif "date" in command:
            date_str = datetime.datetime.now().strftime("%d/%m/%Y")
            speak(f"Today's date is {date_str}")
        elif "search" in command:
            query = command.replace("search for", "")
            speak(f"Here are the search results for {query}")
            webbrowser.open(f"https://www.google.com/search?q={query}")
        elif "wikipedia" in command:
            query = command.replace("wikipedia", "")
            results = wikipedia.summary(query, sentences=2)
            speak("According to Wikipedia")
            speak(results)
        elif "stop" in command or "exit" in command:
            speak("Goodbye!")
            break
        else:
            speak("Sorry, I did not understand that command. Please try again.")

           
```
